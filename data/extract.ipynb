```python
import glob
import os
import numpy as np
import scipy.io as sio
import pandas as pd
```

```python
mat = sio.loadmat('LU_Parameter_Mean_2013_07_01_00_00_00.mat')
colnames = mat['LU_Parameter_mean'][0].dtype.names
print(colnames)
dat = mat['LU_Parameter_mean'][0][0]
print(mat['LU_Parameter_mean'][0][0][colnames.index('Pressure')].flatten())
print(os.getcwd())
for file in glob.glob('./*.mat'):
    fnsplit = file.split('_')
    print(fnsplit[4] + '/' + fnsplit[5] + '/' + fnsplit[3])
```

```python
df = {
        'Grid_Cell':np.empty(0,dtype=np.int32),
        'date':[],
        'T_2m_Avg':np.empty(0),
        'Dwt_2m_Avg':np.empty(0),
        'Ws_10m_Avg':np.empty(0),
        'Wdir_10m_Avg':np.empty(0),
        'Pmsl_Avg':np.empty(0),
        'Veg_Avg':np.empty(0),
        'PM_FRM_mod':np.empty(0)
}
for file in glob.glob('./*.mat'):
    fnsplit = file.split('_')
    date = fnsplit[4] + '/' + fnsplit[5] + '/' + fnsplit[3]
    print(date)
    mat = sio.loadmat(file)
    colnames = mat['LU_Parameter_mean'][0].dtype.names
    dat = mat['LU_Parameter_mean'][0][0]
    
    numGridPoints = dat[0].size
    date = [date]*numGridPoints
    Grid_Cell = np.arange(1,numGridPoints+1)
    T_2m_Avg = dat[colnames.index('T2')].flatten()
    Dwt_2m_Avg = dat[colnames.index('Dew_Point_Temperature')].flatten()
    Ws_10m_Avg = dat[colnames.index('Wind_Speed')].flatten()
    Wdir_10m_Avg = dat[colnames.index('Wind_Direction')].flatten()
    Pmsl_Avg = dat[colnames.index('Pressure')].flatten()
    Veg_Avg = dat[colnames.index('VEGFRA')].flatten()
    PM_FRM_mod = dat[colnames.index('PM25_X')].flatten()
            
    df['Grid_Cell'] = np.append(df['Grid_Cell'], Grid_Cell.astype(np.int32))
    df['date'] += date
    df['T_2m_Avg'] = np.append(df['T_2m_Avg'], T_2m_Avg)
    df['Dwt_2m_Avg'] = np.append(df['Dwt_2m_Avg'], Dwt_2m_Avg)
    df['Ws_10m_Avg'] = np.append(df['Ws_10m_Avg'], Ws_10m_Avg)
    df['Wdir_10m_Avg'] = np.append(df['Wdir_10m_Avg'], Wdir_10m_Avg)
    df['Pmsl_Avg'] = np.append(df['Pmsl_Avg'], Pmsl_Avg)
    df['Veg_Avg'] = np.append(df['Veg_Avg'], Veg_Avg)
    df['PM_FRM_mod'] = np.append(df['PM_FRM_mod'], PM_FRM_mod)

    
df = pd.DataFrame(df)
df.to_csv('LUR.csv', index=False)
```
